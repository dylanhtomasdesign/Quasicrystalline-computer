## 1. The Golden Ratio (φ) and 6D Geometry

The Golden Ratio φ = (1 + √5)/2 ≈ 1.618033988749895 is fundamental to quasicrystalline structures.

### Key Properties:
- **Self-similarity**: φ² = φ + 1
- **Relationship to 5-fold symmetry**: φ appears in pentagonal tessellations
- **6D Origin**: 6-dimensional hypercubic lattices project onto icosahedral quasiperiodic structures

```python
# Golden Ratio and its properties
phi = (1 + np.sqrt(5)) / 2  # Golden Ratio
phi_inv = 1 / phi  # Golden Ratio conjugate

print(f"Golden Ratio φ = {phi:.15f}")
print(f"φ² = {phi**2:.15f}")
print(f"φ + 1 = {phi + 1:.15f}")
print(f"φ - 1 = {phi - 1:.15f}")
print(f"1/φ = {phi_inv:.15f}")
print(f"φ² - φ = {phi**2 - phi:.15f}")
```

```python
# Visualize the Golden Spiral
fig, ax = plt.subplots(figsize=(10, 6))

# Golden spiral using Fibonacci rectangles
def golden_spiral(ax, n_squares=8):
    fibs = [1, 1]
    for i in range(n_squares):
        fibs.append(fibs[-1] + fibs[-2])
    
    colors = list(mcolors.TABLEAU_COLORS.values())[:n_squares]
    
    x, y = 0, 0
    for i in range(n_squares):
        size = fibs[i]
        rect = plt.Rectangle((x, y), size, size, 
                            facecolor=colors[i], 
                            alpha=0.3, edgecolor='black')
        ax.add_patch(rect)
        
        # Position for next square
        if i % 4 == 0:
            y += size
        elif i % 4 == 1:
            x += size
        elif i % 4 == 2:
            y -= fibs[i-1]
        else:
            x -= fibs[i-1]
    
    # Draw spiral arc
    theta = np.linspace(0, 4 * np.pi, 100)
    r = np.exp(theta / np.log(phi))
    ax.plot(r * np.cos(theta), r * np.sin(theta), 'b-', linewidth=2)
    
    ax.set_xlim(-1, 15)
    ax.set_ylim(-1, 12)
    ax.set_aspect('equal')
    ax.set_title('Golden Spiral and Fibonacci Rectangles', fontsize=14)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(True, alpha=0.3)

golden_spiral(ax)
plt.tight_layout()
plt.show()
```

### 6D Geometry: From Hypercubic Lattice to Quasicrystal

The 6D cubic lattice can be projected onto 3D space through an icosahedral window, creating Penrose-like patterns.

```python
# 6D to 3D projection using icosahedral symmetry

# Icosahedral projection vectors (as per Coxeter's construction)
# These are the 6 5-fold axes projected onto 3D
def get_icosahedral_vectors():
    """Return the 6 projection vectors for icosahedral symmetry"""
    tau = phi  # Golden ratio
    vectors = np.array([
        [1, tau, 0],
        [-1, tau, 0],
        [1, -tau, 0],
        [-1, -tau, 0],
        [0, 1, tau],
        [0, -1, tau],
        [0, 1, -tau],
        [0, -1, -tau],
        [tau, 0, 1],
        [tau, 0, -1],
        [-tau, 0, 1],
        [-tau, 0, -1]
    ])
    return vectors / np.linalg.norm(vectors[:3], axis=1).mean()

vectors = get_icosahedral_vectors()
print(f"Icosahedral projection vectors (12 vectors for 6-fold axes):")
print(f"Shape: {vectors.shape}")
print(vectors[:6])
```

```python
# Visualize icosahedral structure
fig = plt.figure(figsize=(12, 5))

# 3D scatter plot of vertices
ax1 = fig.add_subplot(121, projection='3d')

# Generate icosahedron vertices
def icosahedron_vertices():
    phi = (1 + np.sqrt(5)) / 2
    vertices = []
    for i in [-1, 1]:
        for j in [-1, 1]:
            vertices.append([0, i, j*phi])
            vertices.append([i, j*phi, 0])
            vertices.append([i*phi, 0, j])
    return np.array(vertices) / np.sqrt(1 + phi**2)

ico_verts = icosahedron_vertices()
ax1.scatter(ico_verts[:, 0], ico_verts[:, 1], ico_verts[:, 2], 
           c='red', s=100, alpha=0.8)
ax1.set_title('Icosahedron Vertices (6D Projection Basis)')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

# Show projection pattern
ax2 = fig.add_subplot(122)

# Create a quasiperiodic pattern using cut-and-project method
np.random.seed(42)
n_points = 500

# Generate 6D points on a hypercube
points_6d = np.random.uniform(-3, 3, (n_points, 6))

# Project onto 3D using golden ratio weights
weights = np.array([phi**i for i in range(6)])
points_3d = np.dot(points_6d, weights[:3].reshape(-1, 1))

# Filter points within acceptance window
acceptance_radius = 1.5
distances = np.linalg.norm(points_6d[:, 3:], axis=1)
mask = distances < acceptance_radius

ax2.scatter(points_3d[mask, 0], points_3d[mask, 1], 
           c=distances[mask], cmap='viridis', alpha=0.6, s=20)
ax2.set_title('Quasiperiodic Projection (Filtered)')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_aspect('equal')

plt.tight_layout()
plt.show()
```

## 2. Icosahedral Projection

The icosahedral symmetry group has order 60 and provides the mathematical foundation for quasicrystalline computational substrates.

### Key Properties:
- **5-fold rotational symmetry** on multiple axes
- **Quasiperiodicity**: No translational symmetry, but long-range order
- **Self-similarity**: Fractal scaling by φ^n

```python
# Demonstrate icosahedral symmetry operations

def generate_icosahedral_symmetry(n_rotations=12):
    """Generate rotation matrices for icosahedral symmetry"""
    rotations = []
    
    # Basic rotations
    tau = phi
    
    # 72-degree rotations around 5-fold axes
    for i in range(5):
        angle = 2 * np.pi * i / 5
        # Rotation in the plane perpendicular to 5-fold axis
        R = np.array([
            [np.cos(angle), -np.sin(angle), 0],
            [np.sin(angle), np.cos(angle), 0],
            [0, 0, 1]
        ])
        rotations.append(R)
    
    return rotations

rotations = generate_icosahedral_symmetry()
print(f"Generated {len(rotations)} fundamental icosahedral rotations")
print("Each rotation corresponds to 72° rotation around a 5-fold symmetry axis")
```

```python
# Visualize icosahedral tiling with golden ratio scaling
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

def draw_pentagon(ax, center, size, angle=0, color='blue', alpha=0.3):
    """Draw a pentagon at the given center"""
    angles = np.linspace(0, 2*np.pi, 6)[:-1] + angle
    x = center[0] + size * np.cos(angles)
    y = center[1] + size * np.sin(angles)
    ax.fill(x, y, color=color, alpha=alpha, edgecolor='black')

# Create self-similar pentagonal tiling
for idx, scale in enumerate([1, phi, phi**2]):
    ax = axes[idx]
    
    # Central pentagon
    draw_pentagon(ax, (0, 0), 1/scale, color='red', alpha=0.5)
    
    # Surrounding pentagons
    for i in range(5):
        angle = 2 * np.pi * i / 5
        x, y = 1.618/scale * np.cos(angle), 1.618/scale * np.sin(angle)
        draw_pentagon(ax, (x, y), 1/scale, color='blue', alpha=0.3)
    
    ax.set_xlim(-3, 3)
    ax.set_ylim(-3, 3)
    ax.set_aspect('equal')
    ax.set_title(f'Scale: φ^{idx} ≈ {phi**idx:.2f}')
    ax.grid(True, alpha=0.3)

plt.suptitle('Self-Similar Pentagonal Tiling (Golden Ratio Scaling)', fontsize=14)
plt.tight_layout()
plt.show()
```

## 3. Substrate-Problem Alignment Principle

The core innovation of Senemosìa is the **Substrate-Problem Alignment (SPA)** principle:

> *"The optimal computational substrate for a problem is determined by the natural symmetry and dynamics of that problem's solution space."*

### Paradigm Selection Matrix:

| Problem Type | Optimal Paradigm | Key Metric |
|-------------|------------------|------------|
| Sequential Logic | Classical Turing | Clock Speed |
| Geometric/Pattern | 6D Quasicrystalline | Symmetry Match |
| Harmonic/Spectral | Harmonic Resonance | Frequency Alignment |
| Search/Optimization | Quantum | Superposition |
| Sensory/Temporal | Neuromorphic | Temporal Coding |
| Continuous/Physical | Analog/Hybrid | ODE Solver |

```python
# Define the Substrate-Problem Alignment framework

class SubstrateProblemAlignment:
    """
    Framework for aligning computational substrates to problem characteristics.
    """
    
    PARADIGMS = {
        'classical': {
            'symmetry': 'translational',
            'strengths': ['sequential', 'branching', 'logic'],
            'metric': 'MIPS'
        },
        '6d_geometric': {
            'symmetry': 'icosahedral',
            'strengths': ['pattern', 'geometric', 'quasiperiodic'],
            'metric': 'symmetry_coverage'
        },
        'harmonic': {
            'symmetry': 'spectral',
            'strengths': ['resonance', 'frequency', 'wave'],
            'metric': 'Q_factor'
        },
        'quantum': {
            'symmetry': 'superposition',
            'strengths': ['search', 'factoring', 'sampling'],
            'metric': 'qubit_coherence'
        },
        'neuromorphic': {
            'symmetry': 'temporal',
            'strengths': ['spike', 'adaptive', 'sensory'],
            'metric': 'synaptic_events_per_second'
        },
        'analog': {
            'symmetry': 'continuous',
            'strengths': ['ODE', 'optimization', 'physical'],
            'metric': 'precision_equivalent'
        }
    }
    
    def __init__(self):
        self.phi = (1 + np.sqrt(5)) / 2
        
    def compute_alignment_score(self, problem_type, paradigm):
        """Compute alignment score between problem and paradigm."""
        if paradigm not in self.PARADIGMS:
            return 0.0
            
        para = self.PARADIGMS[paradigm]
        
        # Ideal alignments
        alignments = {
            ('sequential', 'classical'): 1.0,
            ('pattern', '6d_geometric'): 1.0,
            ('resonance', 'harmonic'): 1.0,
            ('search', 'quantum'): 1.0,
            ('temporal', 'neuromorphic'): 1.0,
            ('continuous', 'analog'): 1.0,
            
            # Cross-paradigm efficiencies (normalized by φ)
            ('pattern', 'quantum'): 1/self.phi,
            ('temporal', 'neuromorphic'): 1.0,
            ('continuous', 'analog'): 1.0,
        }
        
        return alignments.get((problem_type, paradigm), 0.1)
    
    def select_optimal_substrate(self, problem_types):
        """Select optimal substrate(s) for given problem characteristics."""
        scores = {}
        
        for para in self.PARADIGMS:
            total_score = sum(
                self.compute_alignment_score(pt, para) 
                for pt in problem_types
            )
            scores[para] = total_score / len(problem_types)
        
        return scores

# Demonstrate SPA
spa = SubstrateProblemAlignment()

print("Substrate-Problem Alignment (SPA) Framework")
print("=" * 50)
print("\nExample: Pattern Recognition Problem")
print("-" * 50)

scores = spa.select_optimal_substrate(['pattern', 'temporal'])
for para, score in sorted(scores.items(), key=lambda x: -x[1]):
    bar = "█" * int(score * 20)
    print(f"{para:15s}: {bar} {score:.2%}")
```

## 5. Laplacian Spectral Scheduling

The Laplacian matrix of the quasicrystalline graph provides a spectral basis for task scheduling.

### Key Concept:
Eigenvectors of the graph Laplacian encode natural oscillation modes of the computational substrate, which can be used to optimally assign tasks.

```python
# Laplacian Spectral Scheduling demonstration

import networkx as nx
from scipy import linalg

def create_quasicrystal_graph(size=10):
    """Create a quasiperiodic graph using Penrose-like connectivity"""
    G = nx.Graph()
    
    # Create nodes in quasiperiodic arrangement
    for i in range(size):
        for j in range(size):
            # Golden ratio-based positioning
            x = i + phi * j
            y = j - phi_inv * i
            G.add_node((i, j), pos=(x, y))
    
    # Add edges with quasiperiodic connectivity
    for i in range(size):
        for j in range(size):
            # Connect to nearest neighbors (quasiperiodically)
            for di, dj in [(1, 0), (0, 1), (1, 1), (1, -1)]:
                ni, nj = i + di, j + dj
                if 0 <= ni < size and 0 <= nj < size:
                    G.add_edge((i, j), (ni, nj))
    
    return G

# Create and analyze the graph
np.random.seed(42)
G = create_quasicrystal_graph(8)

# Compute Laplacian spectrum
L = nx.laplacian_matrix(G).toarray()
eigenvalues, eigenvectors = np.linalg.eigh(L)

print(f"Quasicrystal Graph Analysis:")
print(f"  Nodes: {G.number_of_nodes()}")
print(f"  Edges: {G.number_of_edges()}")
print(f"  Laplacian Eigenvalues (first 5): {eigenvalues[:5].round(4)}")
print(f"  Spectral Gap: {eigenvalues[1]:.4f}")
```

```python
# Visualize spectral modes
fig, axes = plt.subplots(2, 3, figsize=(14, 9))
axes = axes.flatten()

pos = nx.get_node_attributes(G, 'pos')

for idx in range(6):
    ax = axes[idx]
    
    if idx == 0:
        # Graph structure
        nx.draw(G, pos, ax=ax, node_size=50, node_color='lightblue',
               edge_color='gray', alpha=0.7)
        ax.set_title(f'Graph Structure')
    else:
        # Spectral mode
        mode = eigenvectors[:, idx]
        node_colors = mode
        nx.draw(G, pos, ax=ax, node_size=50, node_color=node_colors,
               cmap='RdBu', edge_color='gray', alpha=0.8,
               vmin=-np.max(np.abs(mode)), vmax=np.max(np.abs(mode)))
        ax.set_title(f'Eigenmode {idx}\nλ={eigenvalues[idx]:.2f}')

plt.suptitle('Laplacian Spectral Modes of Quasicrystal Graph', fontsize=14)
plt.tight_layout()
plt.show()
```

## 6. Quantum Computing (Grover's Algorithm)

Grover's quantum search algorithm provides a quadratic speedup for unstructured search problems.

### Classical vs Quantum Search Complexity:
- **Classical**: O(N) oracle queries
- **Quantum (Grover)**: O(√N) oracle queries
- **Speedup**: Quadratic improvement

```python
# Grover's Algorithm demonstration

def grovers_algorithm(n_qubits, target_state, n_iterations=None):
    """
    Simplified Grover's Algorithm simulation.
    """
    N = 2 ** n_qubits  # Total states
    
    if n_iterations is None:
        n_iterations = int(np.pi / 4 * np.sqrt(N))
    
    # Initial state |ψ⟩ = |0⟩^{⊗n}
    state = np.zeros(N)
    state[0] = 1.0
    
    # Hadamard to create superposition
    H = np.array([[1, 1], [1, -1]]) / np.sqrt(2)
    H_n = H
    for _ in range(n_qubits - 1):
        H_n = np.kron(H_n, H)
    
    state = H_n @ state
    
    # Oracle and diffusion operators
    oracle = np.eye(N)
    oracle[target_state, target_state] = -1
    
    diffusion = 2 * np.ones((N, N)) / N - np.eye(N)
    
    # Grover iterations
    for _ in range(n_iterations):
        state = oracle @ state
        state = diffusion @ state
    
    # Measure
    probabilities = np.abs(state) ** 2
    measured = np.argmax(probabilities)
    
    return measured, probabilities

# Example: 10-qubit search space
n_qubits = 10
N = 2 ** n_qubits
target = np.random.randint(0, N)

measured, probs = grovers_algorithm(n_qubits, target)

print(f"Grover's Algorithm Simulation ({n_qubits} qubits)")
print("=" * 50)
print(f"Search space size: {N:,} states")
print(f"Target state: {target}")
print(f"Measured result: {measured}")
print(f"Success: {measured == target}")
print(f"Classical queries needed: ~{N}")
print(f"Quantum queries needed: ~{int(np.pi/4*np.sqrt(N))}")
```

```python
# Visualize Grover's amplification

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Probability distribution after Grover's
ax1.bar(range(min(100, N)), probs[:min(100, N)], alpha=0.7, color='steelblue')
ax1.axvline(target, color='red', linestyle='--', linewidth=2, label=f'Target: {target}')
ax1.set_xlabel('State Index')
ax1.set_ylabel('Probability')
ax1.set_title('Grover\'s Algorithm: Amplified Probabilities')
ax1.legend()

# Complexity comparison
n_bits_range = range(4, 16)
classical_complexity = [2**n for n in n_bits_range]
quantum_complexity = [int(np.pi/4 * np.sqrt(2**n)) for n in n_bits_range]

ax2.plot(list(n_bits_range), classical_complexity, 'b-', label='Classical O(N)', linewidth=2)
ax2.plot(list(n_bits_range), quantum_complexity, 'r-', label='Quantum O(√N)', linewidth=2)
ax2.set_yscale('log')
ax2.set_xlabel('Number of Qubits')
ax2.set_ylabel('Queries Required')
ax2.set_title('Quantum Speedup: Grover\'s Algorithm')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 7. Spiking Neural Networks

Neuromorphic computing mimics the brain's spiking neurons for energy-efficient temporal computation.

### Key Features:
- **Temporal coding**: Information in spike timing
- **Leaky Integrate-and-Fire (LIF) neurons**
- **STDP (Spike-Timing-Dependent Plasticity)** for learning

```python
# Leaky Integrate-and-Fire (LIF) Neuron Simulation

class LIFNeuron:
    """Leaky Integrate-and-Fire Neuron Model"""
    
    def __init__(self, tau=20, v_thresh=-50, v_rest=-70, r=10):
        self.tau = tau      # Membrane time constant (ms)
        self.v_thresh = v_thresh  # Threshold potential (mV)
        self.v_rest = v_rest      # Resting potential (mV)
        self.r = r         # Membrane resistance (MΩ)
        self.v = v_rest    # Initial membrane potential
        self.spikes = []
    
    def update(self, i_input, dt=1):
        """Update membrane potential"""
        # Leaky integrate: dv/dt = -(v - v_rest)/tau + i_input
        dv = (-(self.v - self.v_rest) / self.tau + i_input) * dt
        self.v += dv
        
        # Check for spike
        if self.v >= self.v_thresh:
            self.spikes.append(1)
            self.v = self.v_rest  # Reset
        else:
            self.spikes.append(0)
        
        return self.v

# Simulate LIF neuron with different input currents
t = np.arange(0, 200, 1)  # 200 ms

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

input_currents = [1.5, 2.0, 2.5, 3.0]
colors = plt.cm.plasma(np.linspace(0, 0.8, len(input_currents)))

for i_current, current in enumerate(input_currents):
    neuron = LIFNeuron(tau=20)
    voltages = []
    
    for step in t:
        # Constant input current
        v = neuron.update(current, dt=1)
        voltages.append(v)
    
    ax1.plot(t, voltages, color=colors[i_current], 
            label=f'I = {current} mV', alpha=0.8)
    ax2.eventplot([np.where(np.array(neuron.spikes) == 1)[0]], 
                  lineoffsets=i_current+1, linelengths=0.8,
                  colors=[colors[i_current]], label=f'I = {current}')

ax1.axhline(neuron.v_thresh, color='red', linestyle='--', alpha=0.5, label='Threshold')
ax1.axhline(neuron.v_rest, color='green', linestyle=':', alpha=0.5, label='Rest')
ax1.set_xlabel('Time (ms)')
ax1.set_ylabel('Membrane Potential (mV)')
ax1.set_title('Leaky Integrate-and-Fire Neuron Response')
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2.set_xlabel('Time (ms)')
ax2.set_ylabel('Input Current')
ax2.set_title('Spike Trains for Different Input Currents')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
# STDP (Spike-Timing-Dependent Plasticity) Learning Rule

def stdp_weight_change(dt, a_plus=0.01, a_minus=0.012, tau_plus=20, tau_minus=20):
    """
    Compute STDP weight change.
    
    Parameters:
    -----------
    dt : float
        Time difference (pre_time - post_time)
    """
    if dt >= 0:
        # Pre before post: potentiation
        return a_plus * np.exp(-dt / tau_plus)
    else:
        # Post before pre: depression
        return -a_minus * np.exp(dt / tau_minus)  # dt is negative

# Visualize STDP curve
dt_range = np.linspace(-100, 100, 1000)

fig, ax = plt.subplots(figsize=(10, 6))

weights = [stdp_weight_change(dt) for dt in dt_range]

ax.plot(dt_range, weights, 'b-', linewidth=2)
ax.axhline(0, color='gray', linestyle='--', alpha=0.5)
ax.axvline(0, color='gray', linestyle='--', alpha=0.5)

ax.fill_between(dt_range[dt_range >= 0], 0, [weights[i] for i in range(len(dt_range)) if dt_range[i] >= 0],
                alpha=0.3, color='green', label='Potentiation (pre → post)')
ax.fill_between(dt_range[dt_range < 0], 0, [weights[i] for i in range(len(dt_range)) if dt_range[i] < 0],
                alpha=0.3, color='red', label='Depression (post → pre)')

ax.set_xlabel('Δt = t_post - t_pre (ms)', fontsize=12)
ax.set_ylabel('Weight Change (Δw)', fontsize=12)
ax.set_title('STDP Learning Window', fontsize=14)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("STDP Summary:")
print("  • Pre before post (Δt > 0): Weight increases (potentiation)")
print("  • Post before pre (Δt < 0): Weight decreases (depression)")
print("  • Peak potentiation at Δt ≈ 20 ms")
print("  • Asymmetric window: faster depression than potentiation")
```
